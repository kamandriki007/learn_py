```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = x**2
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y, "k-.")
plt.grid()
plt.ylim([0, 26])
plt.xlim([-5,5])
plt.xlabel(r'$x$ [m]', fontsize=18)
plt.ylabel(r'$x^2$ [m]', fontsize=18)
```
