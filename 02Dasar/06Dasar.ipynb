## Functions
Problem: You are given two lists of numbers and you need to find total of each of these list

```python
tom_exp_list = [2100, 3400, 3500]
joe_exp_list = [200, 500, 700]

total=0
for item in tom_exp_list:
    total=total+item
print("Tom's total expenses:",total)

total=0
for item in joe_exp_list:
    total=total+item
print("Joe's total expenses:",total)
```

```python
def calculate_total(exp):
    total = 0
    for item in exp:
        total=total+item
    return total

tom_exp_list = [2100, 3400, 3500]
joe_exp_list = [200, 500, 700]

toms_total=calculate_total(tom_exp_list)
joes_total=calculate_total(joe_exp_list)

print("Tom's total expenses:",toms_total)
print("Joe's total expenses:",joes_total)
```

```python
def sum(a,b):
    total=0
    total=a+b
    return total

n=sum(5,6)
print("Total:",n)
```

```python
def sum(a,b):
    print("a:",a)
    print("b:",b)
    
    total=a+b
    print("Total inside function:",total)
    return total

n=sum(5,6)
print("Total outside function:",n)
```

```python
#Global vs Local Variables
total=0
def sum(a,b):
    print("a:",a)
    print("b:",b)
    
    total=a+b
    print("Total inside function:",total)
    return total

n=sum(b=5,a=6)
print("Total outside function:",total)
```

```python
#Default Arguments
total=0
def sum(a,b=0):
    print("a:",a)
    print("b:",b)
    
    total=a+b
    print("Total inside function:",total)
    return total

n=sum(5)
print("Total outside function:",total)
```
