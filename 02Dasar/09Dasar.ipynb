## JSON
(1) To Create address book and write some records into it
(2) Read this address book

```python
book = {}
book['tom'] = {
    'name': 'tom',
    'address': '1 red street, NY',
    'phone': 98989898
}
book['bob'] = {
    'name': 'bob',
    'address': '1 green street, NY',
    'phone': 23424234
}

import json
s=json.dumps(book)
print(s)
```

```python
book = {}
book['tom'] = {
    'name': 'tom',
    'address': '1 red street, NY',
    'phone': 98989898
}
book['bob'] = {
    'name': 'bob',
    'address': '1 green street, NY',
    'phone': 23424234
}

import json
s=json.dumps(book)
with open("d://learn_py//02Dasar//book.txt","w") as f:
    f.write(s)
```

```python
f=open("d://learn_py//02Dasar//book.txt","r")
s=f.read()
s
```

```python
import json
book=json.loads(s)
book
```
